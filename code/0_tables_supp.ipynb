# Define functions to generate tables

```python
def table_summarize_descriptive(data,var_short,var_long):

        Names = zip(var_short,var_long)

        mean_list = []
        std_list = []
        min_list = []
        max_list = []
        n_list=[]
        variables = []
       

        for i,j in Names: #var_short,var_long,scale
                
                a_mean = data[i].mean()
                a_std  = data[i].std()
                a_min = data[i].min() 
                a_max = data[i].max()
                a_n = data[i].count()

                variables.append(j)
                
                mean_list.append("{:.2f}".format(a_mean))
                std_list.append("{:.2f}".format(a_std))
                min_list.append("{:.2f}".format(a_min))
                max_list.append("{:.2f}".format(a_max))
                n_list.append("{:.0f}".format(a_n))

        df = pd.DataFrame(dict(Variables = variables, Mean=mean_list, SD=std_list, Min=min_list, Max=max_list,N=n_list))

        # return df.to_latex(buf=file,index = False,column_format='lcccc')
        return df
```

```python
def table_summarize_descriptive2(names,data1,vars1,data2,vars2):

        Names = zip(vars1,vars2)

        mean1_list = []
        std1_list = []
        n1_females = []
        n1_males = []
        n1_both = []
        n1_obser = []

        mean2_list = []
        std2_list = []
        n2_females = []
        n2_males = []
        n2_both = []
        n2_obser = []

        variables = []

        std2_list = []


        for var in vars1:

                # print(var)

                mask = data1[var]==1
                data_0 = data1[mask]
               
                n1_females.append("{:.2f}".format(100*data_0['only_females'].mean()))
                n1_males.append("{:.2f}".format(100*data_0['only_males'].mean()))
                n1_both.append("{:.2f}".format(100*data_0['both_genders'].mean()))
                # n1_obser.append("{:.2f}".format(len(data_0)))
                # print(len(data_0))

        for var in vars2:

                mask = data2[var]==1
                data_0=data2[mask]

                n2_females.append("{:.2f}".format(100*data_0['only_females'].mean()))
                n2_males.append("{:.2f}".format(100*data_0['only_males'].mean()))
                n2_both.append("{:.2f}".format(100*data_0['both_genders'].mean()))
                # n2_obser.append("{:.2f}".format(len(data_0)))

        for name in names:

                variables.append(name)

                # print(name)

        for i,j in Names: #var_short,var_long,scale

                # print(i,'',j)
                
                a_mean = data1[i].mean()
                # print(len(data1[i]))
                a_std  = data1[i].std()   
                     

                b_mean = data2[j].mean()
                b_std  = data2[j].std()                
                
                mean1_list.append("{:.3f}".format(a_mean))
                std1_list.append("{:.2f}".format(a_std))
                # n1_obser.append("{:.0f}".format(len(data1))) 
                n1_obser.append("{:.0f}".format(data1[i].sum() )) 

                mean2_list.append("{:.3f}".format(b_mean))
                std2_list.append("{:.2f}".format(b_std))
                # n2_obser.append("{:.0f}".format(len(data2)))   
                n2_obser.append("{:.0f}".format(data2[j].sum() )) 


        df = pd.DataFrame(dict(Variables = variables,  Mean1 = mean1_list, SD1= std1_list,Obs1 =  n1_obser,
                sh1_fem = n1_females, sh1_male = n1_males, sh1_both = n1_both,  Mean2 = mean2_list, SD2= std2_list,
                Obs2 =  n2_obser, sh2_fem = n2_females, sh2_male = n2_males, sh2_both = n2_both))

        # # return df.to_latex(buf=file,index = False,column_format='lccccccc')Mean1=mean1_list, SD1=std1_list,
        return df
```

```python
def table_summarize_descriptive3(data_obs,data_imp,names2,list_obs,list_imp):
        
        Names = zip(list_obs,list_imp)

        mean1_list = []
        std1_list = []
 
        mean2_list = []
        std2_list = []


        variables = []


        for name in names2:

                variables.append(name)

                # print(name)

        for i,j in Names: #var_short,var_long,scale

                # print(i,'',j)

                # (df['minSalaryAnnual'].loc[df['minSalaryAnnual'] != 0]).median()
                
                a_mean = (data_obs[i].loc[data_obs[i] !=0]).mean()
                # print(len(data1[i]))
                a_std  = (data_obs[i].loc[data_obs[i] !=0]).std()  
                     

                b_mean = (data_imp[j].loc[data_imp[j] !=0]).mean()
                b_std  = (data_imp[j].loc[data_imp[j] !=0]).std()             
                
                mean1_list.append("{:.2f}".format(a_mean))
                std1_list.append("{:.2f}".format(a_std))
              
                mean2_list.append("{:.2f}".format(b_mean))
                std2_list.append("{:.2f}".format(b_std))
 


        df = pd.DataFrame(dict(Variables = variables,  Mean1 = mean1_list, SD1= std1_list, Mean2 = mean2_list, SD2= std2_list))

        # # return df.to_latex(buf=file,index = False,column_format='lccccccc')Mean1=mean1_list, SD1=std1_list,
        return df
```

# Table 2: Summary Statistics FK-grade level by JEL code
### Supplemental Materials | page 3

```python
names2 = ["\hspace{3mm}A","\hspace{3mm}B","\hspace{3mm}C","\hspace{3mm}D","\hspace{3mm}E",
        "\hspace{3mm}F","\hspace{3mm}G","\hspace{3mm}H","\hspace{3mm}I","\hspace{3mm}J",
        "\hspace{3mm}K","\hspace{3mm}L","\hspace{3mm}M","\hspace{3mm}N","\hspace{3mm}O",
        "\hspace{3mm}P","\hspace{3mm}Q","\hspace{3mm}R","\hspace{3mm}Z"
        ]

vars1 = ["A_obs","B_obs","C_obs","D_obs","E_obs","F_obs","G_obs","H_obs","I_obs","J_obs",
        "K_obs","L_obs","M_obs","N_obs","O_obs","P_obs","Q_obs","R_obs" ,"Z_obs"]

vars2 = ["A_imp","B_imp","C_imp","D_imp","E_imp","F_imp","G_imp","H_imp","I_imp",
        "J_imp","K_imp","L_imp","M_imp","N_imp","O_imp","P_imp","Q_imp","R_imp","Z_imp"]
```

```python
list_names_obs = []
list_names_imp = []

names = zip(vars1,vars2)

df1 = df1.copy()

for i,j in names:

    nn = i+'FKG'
    mm = j+'FKG'

    list_names_obs.append(nn)
    list_names_imp.append(mm)

    df1.loc[:,nn]=df1[i]*df1['flesch_kincaid_grade_level']
    df.loc[:,mm]=df[j]*df['flesch_kincaid_grade_level']
```

```python
file = '../tables/table2_sm.tex'

table_jel_fkg = table_summarize_descriptive3(df1,df,names2,list_names_obs,list_names_imp)

table_jel_fkg.columns=['Variables','Mean','SD','Mean','SD']

table_jel_fkg.insert(3,' \\multicolumn{1}{l}{}',' ')

table_jel_fkg.to_latex(buf=file, index=False, column_format='lccccc',escape=False)

table_jel_fkg
# &   \multicolumn{2}{c}{Observed JEL codes} &  &\multicolumn{2}{c}{With JEL codes imputed}\\
# \cmidrule(lr){2-3} \cmidrule(lr){5-6} 
```
